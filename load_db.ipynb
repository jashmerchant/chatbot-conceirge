```python
import boto3
import json
import requests
# from decimal import Decimal
access_key=''
secret_access_key=''
```

```python
session=boto3.Session(aws_access_key_id=access_key,aws_secret_access_key=secret_access_key, region_name='us-east-1')
client_dynamo=session.resource('dynamodb')
table=client_dynamo.Table('yelp-restaurant')
```

```python
records=""
with open('chinese-restaurant.json','r') as datafile:
  records=json.load(datafile)
```

```python
print(records)
```

```python
count = 999
for i in records:
    i['serial_no'] = count
    i = json.loads(json.dumps(i), parse_float=Decimal)
    response = table.put_item(Item=i)
    count = count + 1
```
